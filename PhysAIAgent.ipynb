```python
# mcp_bridge_client_repl.py
# pip install openai fastmcp
from __future__ import annotations
import os, json, asyncio, time, logging
from openai import OpenAI
from fastmcp import Client
from fastmcp.client.transports import StreamableHttpTransport
from bridge_session import BridgeSession  # ← 분리된 세션 클래스 임포트

# =========================
# 0) 환경 설정
# =========================
LLM_BASE_URL = os.getenv("LLM_BASE_URL", "http://127.0.0.1:8000/v1")
# 🔁 여기만 Qwen → Llama-3.1-8B-Instruct 로 변경
LLM_MODEL    = os.getenv("LLM_MODEL", "meta-llama/Llama-3.1-8B-Instruct")
LLM_API_KEY  = os.getenv("LLM_API_KEY",  "EMPTY")
MCP_URL      = os.getenv("MCP_URL",     "https://figures-infants-everywhere-updating.trycloudflare.com/mcp")
MAX_TURNS    = int(os.getenv("MAX_TURNS", "8"))
TEMP         = float(os.getenv("TEMP", "0.2"))
MAX_TOKENS   = int(os.getenv("MAX_TOKENS", "512"))
LOG_FILE     = os.getenv("LOG_FILE", "mcp_bridge.log")

# =========================
# 프리셋 질의 (자동 실행 후 종료)
# =========================
'''
PRESET_QUERIES = [
    # --- 카메라/모드/영상 제어 ---
    "열상 모드로 전환",                          # 1
    "주간 모드로 전환",                          # 2
    "SWIR 모드로 전환",                          # 3
    "EO 카메라 3배 확대",                        # 4
    "IR 카메라 5배 확대",                        # 5
    "IR 카메라 흑상 전환",                       # 6
    "IR 카메라 백상 전환",                       # 7
    "주간 카메라 흔들림 보정 시작",              # 8
    "주간 카메라 흔들림 보정 종료",              # 9
    "열상 카메라 흔들림 보정 시작",              # 10
    "열상 카메라 흔들림 보정 종료",              # 11
    "열상 자동초점",                            # 12
    "주간 자동초점",                            # 13

    # --- 팬틸트 제어 ---
    "좌로 20도 회전",                            # 14
    "우로 30도 회전",                            # 15
    "상으로 5도 회전",                           # 16
    "하로 3도 회전",                             # 17
    "방위각 30도로 이동",                       # 18
    "고저각 3도로 이동",                        # 19
    "팬 속도 증가",                             # 20
    "팬 속도 감소",                             # 21
    "틸트 속도 증가",                           # 22
    "틸트 속도 감소",                           # 23
    "정지",                                     # 24

    # --- 전원 제어 ---
    "열상 카메라 전원 켜기",                   # 25
    "열상 카메라 전원 끄기",                   # 26
    "주간 카메라 전원 켜기",                   # 27
    "주간 카메라 전원 끄기",                   # 28
    "LRF 켜기",                                # 29
    "LRF 끄기",                                # 30

    # --- 레이저 거리계(LRF) ---
    "거리 측정 시작",                          # 31
    "타겟 위치 알려줘",                        # 32

    # --- 영상 향상 ---
    "열상 카메라 영상 개선 시작",              # 33
    "열상 카메라 영상 개선 종료",              # 34
    "주간 카메라 영상 개선 시작",              # 35
    "주간 카메라 영상 개선 종료",              # 36

    # --- 위치 이동 ---
    "위도 37°13'56\"N 경도 129°32'25\"E 위치로 이동",  # 37

    # --- 프리셋 위치 이동 ---
    "좌측 빨간 등대 위치로 이동",              # 38
    "우측 방파제 프리셋 위치로 이동",          # 39

    # --- AI 탐지 관련 조회 ---
    "탐지 객체 목록 가져오기",                # 40
    "자동탐지 시작",                           # 41
    "자동탐지 종료",                           # 42
    "자동추적 시작",                           # 43
    "자동추적 종료",                           # 44

    # --- 오토스캔 / 자동감시 ---
    "오토 스캔 목록 보여줘",                  # 45
    "오토 스캔 실행",                          # 46
    "오토 스캔 중지",                          # 47
    "녹화 시작",                               # 48
    "녹화 중지",                               # 49

    # --- 캡처 ---
    "캡처 시작",                               # 50
]

'''
# =========================
# Preset queries (auto execute & exit)
# =========================
'''PRESET_QUERIES = [
    # --- Camera / Mode / Image Control ---
    "Switch to thermal mode",                          # 1
    "Switch to day mode",                              # 2
    "Switch to SWIR mode",                             # 3
    "Zoom EO camera to 3x",                            # 4
    "Zoom IR camera to 5x",                            # 5
    "Switch IR camera to black-hot",                   # 6
    "Switch IR camera to white-hot",                   # 7
    "Start day camera stabilization",                  # 8
    "Stop day camera stabilization",                   # 9
    "Start thermal camera stabilization",              # 10
    "Stop thermal camera stabilization",               # 11
    "Thermal auto focus",                              # 12
    "Day auto focus",                                  # 13

    # --- Pan / Tilt Control ---
    "Pan left 20 degrees",                             # 14
    "Pan right 30 degrees",                            # 15
    "Tilt up 5 degrees",                               # 16
    "Tilt down 3 degrees",                             # 17
    "Move to azimuth 30 degrees",                      # 18
    "Move to elevation 3 degrees",                     # 19
    "Increase pan speed",                              # 20
    "Decrease pan speed",                              # 21
    "Increase tilt speed",                             # 22
    "Decrease tilt speed",                             # 23
    "Stop",                                            # 24

    # --- Power Control ---
    "Power on thermal camera",                         # 25
    "Power off thermal camera",                        # 26
    "Power on day camera",                             # 27
    "Power off day camera",                            # 28
    "Turn LRF on",                                     # 29
    "Turn LRF off",                                    # 30

    # --- Laser Range Finder (LRF) ---
    "Start distance measurement",                      # 31
    "Show target position",                            # 32

    # --- Image Enhancement ---
    "Start thermal image enhancement",                 # 33
    "Stop thermal image enhancement",                  # 34
    "Start day image enhancement",                     # 35
    "Stop day image enhancement",                      # 36

    # --- Position Move ---
    "Move to latitude 37°13'56\"N longitude 129°32'25\"E",  # 37

    # --- Preset Position Move ---
    "Move to left red lighthouse preset position",     # 38
    "Move to right breakwater preset position",        # 39

    # --- AI Detection / Query ---
    "Get detected object list",                        # 40
    "Start auto detection",                            # 41
    "Stop auto detection",                             # 42
    "Start auto tracking",                             # 43
    "Stop auto tracking",                              # 44

    # --- Auto Scan / Surveillance ---
    "Show auto scan list",                             # 45
    "Start auto scan",                                 # 46
    "Stop auto scan",                                  # 47
    "Start recording",                                 # 48
    "Stop recording",                                  # 49

    # --- Capture ---
    "Start capture",                                   # 50
]
'''

'''PRESET_QUERIES = [
    "열상 모드로 전환 후 IR 카메라 5배 확대",              # 1
    "주간 모드로 전환 후 EO 카메라 3배 확대",              # 2
    "주간 카메라 흔들림 보정 시작 후 주간 자동초점 실행",  # 3
    "열상 카메라 흔들림 보정 시작 후 열상 자동초점 실행",  # 4
    "좌로 20도 회전 후 상으로 5도 회전",                   # 5
    "우로 30도 회전 후 하로 3도 회전",                     # 6
    "방위각 30도로 이동 후 고저각 3도로 이동",             # 7
    "팬 속도 증가 후 틸트 속도 증가",                       # 8
    "열상 카메라 전원 켜기 후 열상 모드로 전환",           # 9
    "주간 카메라 전원 켜기 후 주간 모드로 전환",           # 10
    "LRF 켜기 후 거리 측정 시작",                           # 11
    "거리 측정 시작 후 타겟 위치 알려줘",                   # 12
    "열상 카메라 영상 개선 시작 후 IR 카메라 흑상 전환",   # 13
    "주간 카메라 영상 개선 시작 후 EO 카메라 3배 확대",    # 14
    "좌측 빨간 등대 위치로 이동 후 자동탐지 시작",         # 15
    "우측 방파제 프리셋 위치로 이동 후 자동추적 시작",     # 16
    "오토 스캔 목록 보여줘 후 오토 스캔 실행",              # 17
    "오토 스캔 실행 후 녹화 시작",                          # 18
    "녹화 시작 후 캡처 시작",                               # 19
    "자동탐지 시작 후 탐지 객체 목록 가져오기",             # 20
]
'''

'''
PRESET_QUERIES = [
    "Switch to thermal mode then zoom IR camera to 5x",              # 1
    "Switch to day mode then zoom EO camera to 3x",                  # 2
    "Start day camera stabilization then run day autofocus",         # 3
    "Start thermal camera stabilization then run thermal autofocus", # 4
    "Rotate left 20 degrees then tilt up 5 degrees",                 # 5
    "Rotate right 30 degrees then tilt down 3 degrees",              # 6
    "Move to azimuth 30 degrees then move to elevation 3 degrees",   # 7
    "Increase pan speed then increase tilt speed",                   # 8
    "Turn on thermal camera then switch to thermal mode",            # 9
    "Turn on day camera then switch to day mode",                    # 10
    "Turn on LRF then start distance measurement",                   # 11
    "Start distance measurement then show target position",          # 12
    "Start thermal image enhancement then switch IR to black-hot",   # 13
    "Start day image enhancement then zoom EO camera to 3x",         # 14
    "Move to left red lighthouse preset then start auto detection",  # 15
    "Move to right breakwater preset then start auto tracking",      # 16
    "Show auto scan list then start auto scan",                      # 17
    "Start auto scan then start recording",                          # 18
    "Start recording then start capture",                            # 19
    "Start auto detection then get detected object list",            # 20
]
'''

'''
PRESET_QUERIES = [
    "좌측 빨간 등대 근처에 선박이 있으면 자동으로 찾아서 계속 추적해줘.", # O

    "해안 전체를 스캔하고 선박을 자동으로 집중 추적해줘.", # O

    "이 구역을 계속 감시하다가 객체가 나타나면 즉시 알려줘.", # O

    "방파제 근처의 표적을 찾아 정확한 좌표를 알려줘.", # O

    "전체 감시를 수행하고 가장 의심스러운 표적을 지속적으로 추적해줘.", # X

    "현재 화면 내 여러 객체가 있는지 확인하고 가장 가까운 객체를 우선적으로 관측해줘.", # X

    "장거리 감시를 수행하고 필요할 경우 영상 흔들림을 자동으로 보정해줘.", # X

    "이 구역에서 표적을 탐색하고 가장 먼저 발견된 대상을 계속 추적해줘.", # O

    "관심 객체가 탐지되면 자동으로 캡처를 수행해줘.", # X

    "등대 구역을 순찰하면서 움직임이 감지되면 자동으로 녹화를 시작해줘.", # X
]
'''
PRESET_QUERIES = [
    "If there is any vessel near the left red lighthouse, automatically find it and keep tracking it.",  # O

    "Scan the entire coastline and automatically focus tracking on any vessel you detect.",  # O

    "Continuously monitor this area and notify me immediately when any object appears.",  # O

    "Find a target near the breakwater and report its exact coordinates.",  # O

    "Perform full-area surveillance and continuously track the most suspicious target.",  # X

    "Check if there are multiple objects in the current view and prioritize observing the closest one.",  # X

    "Perform long-range surveillance and automatically stabilize the image when necessary.",  # X

    "Search for targets in this area and keep tracking the first one you detect.",  # O

    "When an object of interest is detected, automatically capture a snapshot.",  # X

    "Patrol the lighthouse area and automatically start recording when movement is detected.",  # X
]

# =========================
# 로거 설정 (파일로만 기록)
# =========================
logger = logging.getLogger("mcp_bridge")
logger.setLevel(logging.DEBUG)

# 1) 중복/기존 핸들러 제거
logger.handlers.clear()

# 2) 파일 핸들러만 부착
_fh = logging.FileHandler(LOG_FILE, encoding="utf-8")
_fh.setLevel(logging.DEBUG)
_fh.setFormatter(logging.Formatter(
    "[%(asctime)s] %(levelname)s %(message)s",
    datefmt="%H:%M:%S"
))
logger.addHandler(_fh)

# 3) 루트로 전파 금지 → 콘솔로 새어 나가는 것 차단
logger.propagate = False

# (선택) 코랩/주피터 루트 콘솔 핸들러 제거 + 루트 로그 레벨 상향
root = logging.getLogger()
for h in root.handlers[:]:
    if isinstance(h, logging.StreamHandler):
        root.removeHandler(h)
root.setLevel(logging.WARNING)

# (선택) 서드파티 로거 소음 억제
for noisy in ("openai", "httpx", "fastmcp", "urllib3", "asyncio"):
    logging.getLogger(noisy).setLevel(logging.WARNING)
    logging.getLogger(noisy).propagate = False

# =========================
# 프리셋 실행 루프 (REPL 없음)
# =========================
async def run_repl():
    logger.info("===== SETUP =====")
    logger.info(f"[ENV] MCP_URL={MCP_URL}")
    logger.info(f"[ENV] LLM_BASE_URL={LLM_BASE_URL}")
    logger.info(f"[ENV] LLM_MODEL={LLM_MODEL}")
    logger.info(f"[ENV] TEMP={TEMP}, MAX_TOKENS={MAX_TOKENS}, MAX_TURNS={MAX_TURNS}")

    transport = StreamableHttpTransport(url=MCP_URL)
    mcp = Client(transport)
    async with mcp:
        t0 = time.perf_counter()
        await mcp.ping()
        logger.info(f"[MCP] ping OK ({(time.perf_counter()-t0):.2f}s)")

        mcp_tools = await mcp.list_tools()
        logger.info(f"[MCP] tools count = {len(mcp_tools)}")
        for t in mcp_tools[:30]:
            name = getattr(t, "name", None) or t.get("name")
            logger.info(f"  - {name}")
        if len(mcp_tools) > 30:
            logger.info(f"  ... (+{len(mcp_tools)-30} more)")

        # LLM 준비
        llm = OpenAI(base_url=LLM_BASE_URL, api_key=LLM_API_KEY)

        # 세션 생성 (도구 목록 메모리에 유지)
        sess = BridgeSession(
            llm=llm,
            mcp=mcp,
            mcp_tools=mcp_tools,
            temp=0.2,
            max_tokens=512,
            max_turns=3,
            language="en",   # 🔹 지금 상황
        )


        logger.info("MCP Bridge preset runner. (no interactive REPL)")
        logger.info(f"Preset query count = {len(PRESET_QUERIES)}")

        # =========================
        # 프리셋 질의 자동 실행 후 종료
        # =========================
        for q in PRESET_QUERIES:
            #EO 3배 확대, IR 흑상, 좌로 20도 … 이 서로 기억을 공유할 필요가 거의 없음.
            #오히려, 앞에서 “이미 EO 3배 확대/회전/방위각 했다”라는 문장이 히스토리에 남아 있으니까,
            #모델 입장에서는 “아까 다 했던 얘기니까 그냥 말로만 이어가자”라고 착각할 여지가 생깁니다.
            sess.reset()
            print(f"you> {q}")
            try:
                result = await sess.handle_one_turn(q, model=LLM_MODEL, debug=True)

                print("[AIAGENT DEBUG]")
                print(json.dumps(result, ensure_ascii=False, indent=2))

                # 도구 사용 여부 및 에러 여부 판단
                turns = result.get("turns") or []

                # 1) tool_calls가 한 번이라도 있으면 도구 사용
                used_any_tool = any(
                    (t.get("tool_calls") for t in turns)
                )

                # 2) tool_results 중 ok=False 가 하나라도 있으면 에러로 간주
                has_tool_error = False
                for t in turns:
                    for tr in t.get("tool_results") or []:
                        r = tr.get("result") or {}
                        if not r.get("ok", False):
                            has_tool_error = True
                            break
                    if has_tool_error:
                        break

                if (not used_any_tool) or has_tool_error:
                    final_msg = "MCP 도구 호출에 실패했습니다. (자세한 내용은 로그를 확인해 주세요.)"
                else:
                    final_msg = result.get("final_answer")

                print("[AIAGENT FINAL]", final_msg)
                print()
                print()

            except Exception as e:
                # 로그 + 콘솔 둘 다 에러 출력
                logger.exception(f"[PRESET] handle_one_turn error for '{q}': {e}")
                print(f"[AIAGENT ERROR] {type(e).__name__}: {e}")

        logger.info("All preset queries finished. Exiting.")
        return

# =========================
# 엔트리포인트
# =========================
if __name__ == "__main__":
    try:
        try:
            running = asyncio.get_running_loop()
        except RuntimeError:
            running = None

        if running and running.is_running():
            # 노트북/코랩 환경: 바로 await 실행
            logger.info("[RUN] Notebook/Jupyter loop detected → awaiting run_repl() directly")
            await run_repl()
        else:
            asyncio.run(run_repl())
    except KeyboardInterrupt:
        logger.info("KeyboardInterrupt: exit.")
```
